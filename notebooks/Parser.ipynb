```python
import pandas as pd
import numpy as np
```

# Предобработка txt

```python
f = open('jokes2.txt', 'r')
l = [line.strip() for line in f]
```

```python
start = 0
jokes = []
for i in range(len(l)):
    if l[i] == '##########':
        jokes.append(l[start:i])
        start = i+1
```

```python
final_jokes =[]
for joke in jokes:
    final_jokes.append(' '.join(joke)) 
```

# Предобработка csv 

```python
data = pd.read_csv('jokes.csv', sep='\t', header = None)
```

```python
start = 0
jokes = []
for i in range(data.shape[0]):
    if data[data.index == i].values[0][0] == '##########':
        jokes.append([element[0] for element in data.values[start:i].tolist()])
        start = i+1
```

```python
final_jokes =[]
for joke in jokes:
    final_jokes.append(' '.join(joke)) 
```

# Сохранение и загрузка массива шуток

```python
def saveList(myList,filename):
    np.save(filename,myList)
```

```python
def losdList(filename):
    tempNumpyArray = np.load(filename)
    return tempNumpyArray.tolist()
```
